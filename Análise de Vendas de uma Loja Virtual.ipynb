```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
# Ajustando o estilo dos gráficos
plt.style.use('seaborn-darkgrid')  # Estilo de grade com fundo claro
sns.set_palette('deep')  # Paleta de cores padrão, agradável para visualizações
```

```python
# Carregando o arquivo CSV
df = pd.read_csv('dataset_vendas_eletronicos.csv')
```

```python
# Visualizando as primeiras linhas do dataset
df.head()
```

```python
# Informações gerais do DataFrame
df.info()
```

```python
# Verificando valores ausentes
print("\nValores nulos por coluna:")
print(df.isnull().sum())
```

```python
# Verificando duplicatas
print(f"\nLinhas duplicadas: {df.duplicated().sum()}")
```

```python
# Convertendo a coluna de data, tratar como data (datetime) no pandas e não como texto (string).
df['Data da Venda'] = pd.to_datetime(df['Data da Venda'])
```

```python
# Criando coluna de Receita Total
df['Receita Total'] = df['Preço Unitário'] * df['Quantidade']
```

```python
# Extraindo colunas auxiliares
df['Ano'] = df['Data da Venda'].dt.year
df['Mês'] = df['Data da Venda'].dt.month
df['Dia da Semana'] = df['Data da Venda'].dt.day_name()
```

```python
# Visualizando o resultado
df.head()
```

```python
# Total de Vendas por Mês: como as vendas variam ao longo dos meses

# Agrupando por mês e somando a receita
vendas_por_mes = df.groupby('Mês')['Receita Total'].sum().sort_index()
```

```python
# Gráfico
plt.figure(figsize=(10, 5))
sns.barplot(x=vendas_por_mes.index, y=vendas_por_mes.values)
plt.title('Total de Vendas por Mês')
plt.xlabel('Mês')
plt.ylabel('Receita Total (R$)')
plt.show()
```

```python
# Produtos Mais Vendidos (em quantidade)

# Agrupando por produto e somando quantidades
mais_vendidos = df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=False)
```

```python
# Gráfico
plt.figure(figsize=(10, 5))
sns.barplot(x=mais_vendidos.index, y=mais_vendidos.values)
plt.title('Produtos Mais Vendidos (Quantidade)')
plt.xlabel('Produto')
plt.ylabel('Quantidade Vendida')
plt.xticks(rotation=45)
plt.show()
```

```python
# Faturamento por Estado

# Receita por estado
vendas_por_estado = df.groupby('Estado')['Receita Total'].sum().sort_values(ascending=False)
```

```python
# Gráfico
plt.figure(figsize=(10, 5))
sns.barplot(x=vendas_por_estado.index, y=vendas_por_estado.values)
plt.title('Faturamento por Estado')
plt.xlabel('Estado')
plt.ylabel('Receita Total (R$)')
plt.show()
```

```python
# Distribuição das Formas de Pagamento

# Contagem das formas de pagamento
formas_pagamento = df['Forma de Pagamento'].value_counts()
```

```python
# Gráfico
plt.figure(figsize=(6, 6))
plt.pie(formas_pagamento, labels=formas_pagamento.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribuição das Formas de Pagamento')
plt.axis('equal')
plt.show()
```

```python
# Vendas por Dia da Semana

# Receita por dia da semana
vendas_por_dia = df.groupby('Dia da Semana')['Receita Total'].sum()
```

```python
# Para garantir a ordem dos dias
dias_ordem = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
vendas_por_dia = vendas_por_dia.reindex(dias_ordem)
```

```python
# Gráfico
plt.figure(figsize=(10, 5))
sns.barplot(x=vendas_por_dia.index, y=vendas_por_dia.values)
plt.title('Receita por Dia da Semana')
plt.xlabel('Dia da Semana')
plt.ylabel('Receita Total (R$)')
plt.xticks(rotation=45)
plt.show()
```
